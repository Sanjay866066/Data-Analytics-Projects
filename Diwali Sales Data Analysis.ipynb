```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
%matplotlib inline
import seaborn as sea
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df["Amount"] = df["Amount"]. astype("int")
```

# Exploratory Data Analysis

```python
df.columns
```

```python
ab = sea.countplot(data = df , x = "Gender",hue = 'Gender')
for bars in ab.containers:
    ab.bar_label(bars)
    
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sea.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, hue = 'Gender')
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

# AGE

```python
df.columns
```

```python
sea.countplot(data = df, x = 'Age Group', hue = 'Age Group')
```

```python
ax = sea.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sea.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age, hue = 'Age Group')
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

# State

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(15)

sea.set(rc={'figure.figsize':(23,7)})
sea.barplot(data = sales_state, x = 'State',y= 'Orders', hue = 'State' )
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


# Marital Status

```python
df.columns
```

```python
ax = sea.countplot(data = df, x = 'Marital_Status')

sea.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sea.set(rc={'figure.figsize':(10,5)})
sea.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

# Occupation

```python
sea.set(rc={'figure.figsize':(20,5)})
ax = sea.countplot(data = df, x = 'Occupation', hue = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sea.set(rc={'figure.figsize':(20,5)})
sea.barplot(data = sales_state, x = 'Occupation',y= 'Amount', hue = 'Occupation')
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

# Product Category

```python
df.columns
```

```python
sea.set(rc={'figure.figsize':(25,5)})
ax = sea.countplot(data = df, x = 'Product_Category', hue = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sea.set(rc={'figure.figsize':(20,5)})
sea.barplot(data = sales_state, x = 'Product_Category',y= 'Amount', hue = 'Product_Category')
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sea.set(rc={'figure.figsize':(20,5)})
sea.barplot(data = sales_state, x = 'Product_ID',y= 'Orders', hue = 'Product_ID')
```

```python
fig1, ax1 = mp.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
